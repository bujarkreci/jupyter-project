```python
#replace part of string
import pandas as pd
import re

def remove_whitespace(x):    
    try:
        # Remove spaces inside of the string
        x = "".join(x.split())

    except:
        pass
    return x

ex3 = res.copy()
#in lower case the whole
ex3["NationalTeamNameClean"] = ex3['NationalTeamName'].str.lower()

#remove whitespace in the beginning and in the end (left and right)
ex3["NationalTeamNameClean"] = ex3['NationalTeamNameClean'].str.strip()

#remove only white space everywhere in the column
ex3["NationalTeamNameClean"] = ex3["NationalTeamNameClean"].apply(remove_whitespace)

# Replace single substring by adding new column
#ex3["replaced"] = ex3["NationalTeamNameClean"].replace('usa','unitedstates', regex=True)

# Replace single substring in the same column
ex3["NationalTeamNameClean"] = ex3["NationalTeamNameClean"].replace('usa','unitedstates', regex=True)
ex3.tail(20)
```

```python
import xlsxwriter
import numpy as np
import os

def append_xls(fpath, l, df, sheet_name):
    workbook = xlsxwriter.Workbook(fpath)
    worksheet = workbook.add_worksheet(sheet_name)
    
    text_format = workbook.add_format({'align': 'left', 'bold': True})
    header_format = workbook.add_format({'align': 'left', 'fg_color': 'gray', 'bold': True})

    # Widen the first column to make the text clearer.
    #worksheet.set_column('A:A', 30)
    #cols = ["Team name","FIFA code","Captain","Head coach","Confederation"]
    # Insert an image.
    worksheet.write_row(1, 1, l.columns, header_format)    
    for i, colu in enumerate(l.columns, start=1):
        worksheet.write_row(2, i, l[colu], text_format)
    #worksheet.insert_image('G2', 'python.png')
    
    worksheet.set_column(9, 10)
    #worksheet.set_row(1, 7)
    
    worksheet.write_row(3, 1, df.columns, header_format)
    numpimg = np.array(df["FlagPath"])
    start_row = 3
    for i, column in enumerate(df.columns, start=1):
        worksheet.write_column(start_row+1, i, df[column], text_format)
    
    image_width = 204.0
    image_height = 120.0

    cell_width = 54.0
    cell_height = 26.0

    x_scale = cell_width/image_width
    y_scale = cell_height/image_height
    
    for ima in numpimg:
        worksheet.insert_image(start_row+1, 8, ima, {'x_scale': x_scale, 'y_scale': y_scale})
        worksheet.set_row(start_row+1, 25)
        start_row +=1
    
    #worksheet.write_row('A2', l)
    # Insert an image offset in the cell.
    #worksheet.write('A4', df)
    #worksheet.insert_image('G4', df["FlagPath"], {'x_offset': 5, 'y_offset': 5})

    # Insert an image with scaling.
    #worksheet.write('A23', 'Insert a scaled image:')
    #worksheet.insert_image('B23', 'python.png', {'x_scale': 0.5, 'y_scale': 0.5})

    workbook.close()
        
cols = ["Position","Player name","Date of birth","Caps","Goals","Current club","Country","FlagPath"]
CompleteShow = CompleteShow[cols]
ass=dict(team[0])
ass = ass['team head']
ass = pd.DataFrame(ass)
ass['Team name'] = team[0].teamname
ass = ass.loc[:, ass.columns != 'teamid']

##teamname = pd.DataFrame(team[1].teamname)
#dfNew=pd.concat([ass,CompleteShow],axis=0, ignore_index=True)
#append_to_excel('FifaWC2022/filesexport.xlsx', dfNew, team[0].teamname)

append_xls('FifaWC2022/filesexport.xlsx',ass, CompleteShow, team[0].teamname)
ass
```
