```python
import pandas as pd
import numpy as np
import re

df = pd.DataFrame({'col': {0: '2020-12-08', 
                           1: '2020-12-08', 
                           2: '2020-12-08', 
                           3: '2020-12-08', 
                           4: '2020-12-08', 
                           5: '2020-12-08', 
                           6: '2020-12-08', 
                           7: '2020-12-08', 
                           8: '2020-12-08', 
                           9: '2020-12-08'}})

#option 1
df['col_substring'] = df['col'].str[:4]

#option 2
df['col_substring'] = df['col'].str.slice(0, 4)

#option 3
df['col_substring'] = df['col'].apply(lambda x: x[:4])

#option 4
df['col_substring'] = df['col'].apply(lambda x: re.findall(r'^.{4}', x)[0])

#option 5
df['col_substring'] = np.vectorize(lambda x: x[:4])(df['col'])
```

```python
#replace part of string
import pandas as pd
import re

def remove_whitespace(x):    
    try:
        # Remove spaces inside of the string
        x = "".join(x.split())

    except:
        pass
    return x

ex3 = res.copy()
#in lower case the whole
ex3["NationalTeamNameClean"] = ex3['NationalTeamName'].str.lower()

#remove whitespace in the beginning and in the end (left and right)
ex3["NationalTeamNameClean"] = ex3['NationalTeamNameClean'].str.strip()

#remove only white space everywhere in the column
ex3["NationalTeamNameClean"] = ex3["NationalTeamNameClean"].apply(remove_whitespace)

# Replace single substring by adding new column
#ex3["replaced"] = ex3["NationalTeamNameClean"].replace('usa','unitedstates', regex=True)

# Replace single substring in the same column
ex3["NationalTeamNameClean"] = ex3["NationalTeamNameClean"].replace('usa','unitedstates', regex=True)
ex3.tail(20)
```

```python
# Replace multiple substrings
ex4 = res.copy()
ex4["NationalTeamNameClean"] = ex4['NationalTeamName'].str.lower()
ex4["NationalTeamNameClean"] = ex4['NationalTeamNameClean'].str.strip()

#remove all whitespaces, special characters that are not numbers or letters
ex4["NationalTeamNameClean"] = ex4["NationalTeamNameClean"].str.replace('&', 'and', regex=True)
ex4["NationalTeamNameClean"] = ex4["NationalTeamNameClean"].str.replace('\W', '', regex=True)
ex4.iloc[40:60]
```
